```python
import pandas as pd

# a) Create the DataFrame for the given table.
data = {
    'Item': ['TV', 'TV', 'TV', 'AC'],
    'Company': ['LG', 'VIDEOCON', 'LG', 'SONY'],
    'Rupees': [12000, 10000, 15000, 14000],
    'USD': [700, 650, 800, 750]
}

Product = pd.DataFrame(data)
print("Initial DataFrame:")
print(Product)

# b) Add new rows to the DataFrame.
new_data = pd.DataFrame([
    {'Item': 'AC', 'Company': 'SONY', 'Rupees': 16000, 'USD': 900},
    {'Item': 'TV', 'Company': 'SAMSUNG', 'Rupees': 17000, 'USD': 950}
])

Product = pd.concat([Product, new_data], ignore_index=True)
print("\nDataFrame after adding new rows:")
print(Product)

# c) Display the maximum price of LG TV.
lg_tv = Product[(Product['Item'] == 'TV') & (Product['Company'] == 'LG')]
max_price_lg_tv = lg_tv['Rupees'].max()
print("\nMaximum price of LG TV:", max_price_lg_tv)

# d) Display the sum of all product prices (Rupees).
total_price = Product['Rupees'].sum()
print("\nTotal sum of all product prices (Rupees):", total_price)

# e) Display the median of the USD of Sony products.
sony_products = Product[Product['Company'] == 'SONY']
median_usd_sony = sony_products['USD'].median()
print("\nMedian USD value of SONY products:", median_usd_sony)
```

```python
import pandas as pd

# a) To create the above DataFrame.
data = {
    'Name': ['Aparna', 'Pankaj', 'Ram', 'Ramesh', 'Naveen', 'Krrishnav', 'Bhawna'],
    'Degree': ['MBA', 'BCA', 'M.Tech', 'MBA', None, 'BCA', 'MBA'],
    'Score': [90.0, None, 80.0, 98.0, 97.0, 78.0, 89.0]
}

df = pd.DataFrame(data)
print("Original DataFrame:")
print(df)

# b) To print the Degree and maximum marks in each stream.
print("\nMaximum Score per Degree:")
print(df.groupby('Degree')['Score'].max())

# c) To fill the NaN values with 76.
df_filled = df.fillna(76)
print("\nDataFrame after filling NaN with 76:")
print(df_filled)

# d) To set the index to Name.
df_indexed = df.set_index('Name')
print("\nDataFrame with Name as index:")
print(df_indexed)

# e) To display the name and degree-wise average marks of each student.
print("\nAverage Score by Name and Degree:")
print(df.groupby(['Name', 'Degree'])['Score'].mean())

# f) To count the number of students in MBA.
mba_count = df[df['Degree'] == 'MBA'].shape[0]
print("\nNumber of students in MBA:", mba_count)

# g) To print the mode marks of BCA students.
bca_scores = df[df['Degree'] == 'BCA']['Score']
print("\nMode score of BCA students:", bca_scores.mode().values)
```

```python
import numpy as np

df['Loan_Status'] = np.random.choice([0, 1], size=len(df))

from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(
    df.drop(columns=['Loan_Status']),
    df['Loan_Status'],
    test_size=0.3,
    random_state=0
)
```

```python
constant_features = [
    feature for feature in x_train.select_dtypes(include=['number']).columns
    if x_train[feature].nunique() == 1
]

print(f"Number of constant numeric features: {len(constant_features)}")
```

```python
x_train.drop(labels=constant_features, axis=1, inplace=True)
x_test.drop(labels=constant_features, axis=1, inplace=True)
```

```python
print(df.columns.tolist())
```

```python
import numpy as np

quasi_constant_feat = []

for feature in x_train.columns:

    predominant = (x_train[feature].value_counts() / np.float64(len(x_train))) \
                    .sort_values(ascending=False).values[0]

    if predominant > 0.999:
        quasi_constant_feat.append(feature)

print(f"Number of quasi-constant features: {len(quasi_constant_feat)}")
print(quasi_constant_feat)
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(
    df.drop(labels=['Loan_Status'], axis=1),
    df['Loan_Status'],
    test_size=0.3,
    random_state=0
)
```

```python
duplicated_feat = []

for i in range(0, len(x_train.columns)):

    col_1 = x_train.columns[i]

    for col_2 in x_train.columns[i + 1:]:

        if x_train[col_1].equals(x_train[col_2]):
            duplicated_feat.append(col_2)

duplicated_features = set(duplicated_feat)

print(f"Number of duplicated features: {len(duplicated_features)}")
print(duplicated_features)
```

```python
x_train.drop(labels=quasi_constant_feat, axis=1, inplace=True)
x_test.drop(labels=quasi_constant_feat, axis=1, inplace=True)

print(x_train.shape, x_test.shape)
```

```python
grouped_feature_ls = []
correlated_groups = []

def correlation(dataset, threshold):
    col_corr = set()
    numeric_dataset = dataset.select_dtypes(include=np.number)
    corr_matrix = numeric_dataset.corr()

    for i in range(len(corr_matrix.columns)):
        for j in range(i):

            if abs(corr_matrix.iloc[i, j]) > threshold:
                colname = corr_matrix.columns[i]
                col_corr.add(colname)

    return col_corr

corr_features = correlation(x_train, 0.9)
correlated_features = set(corr_features)
print(f"Number of correlated features to drop: {len(correlated_features)}")
```

```python
# Dropping correlated features
x_train.drop(labels=correlated_features, axis=1, inplace=True)
x_test.drop(labels=correlated_features, axis=1, inplace=True)

# Checking the shapes
x_train.shape, x_test.shape
```

```python
from sklearn.feature_selection import mutual_info_classif
from sklearn.preprocessing import LabelEncoder
import pandas as pd
import matplotlib.pyplot as plt

x_encoded = x_train.copy()

for col in x_encoded.select_dtypes(include='object').columns:
    le = LabelEncoder()
    x_encoded[col] = le.fit_transform(x_encoded[col].astype(str))

x_encoded = x_encoded.fillna(0)

mi = mutual_info_classif(x_encoded, y_train)

mi_series = pd.Series(mi, index=x_encoded.columns)
mi_sorted = mi_series.sort_values(ascending=False)

plt.figure(figsize=(20, 8))
mi_sorted.plot(kind='bar')
plt.title('Mutual Information of Features')
plt.ylabel('MI Score')
plt.xlabel('Features')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
